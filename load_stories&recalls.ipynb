```python
import os
import glob
import numpy as np
import pandas as pd
import json

dr = os.getcwd()
story_names = np.array(['schissel_v1-pool', 'schissel_v2-lake', 'boyscout',
       'triplett_v2-catlady', 'triplett_v1-rookie', 'hester_v1-park',
       'doSomething', 'hester_v2-church', 'stein', 'bear', 'youngest'],
      dtype='<U32')
```

```python
# load the segmented stories
story_name = story_names[0]
story_path = dr + "/data/Stories_and_recalls/" + story_name + ".json"
with open(story_path, "r", encoding="utf-8") as f:
    story_json = json.load(f)

print('Story name:', story_name)
print(story_json['segmentation'])
```

```python
# load the first 10 recalls
subject_inds = list(story_json["recalls segmented"].items())[:10]

for subject_id, recall_text in subject_inds:
    print(f"Segmented recall {subject_id} for {story_name}")
    print(recall_text)
```
